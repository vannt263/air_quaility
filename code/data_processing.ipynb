```python
folder_path = "../Mua gio 31 tram do tai Da Nang tu 11-10 den 19-10"

data = []
idx = 1
for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    if idx == 1:
        subdata = pd.read_excel(file_path, header=3, usecols='A, C:Z')
    else:
        subdata = pd.read_excel(file_path, header=3, usecols='C:Z')
    idx+=1
    data.append(subdata)
```

```python
final_data = pd.concat(data, axis=1)
```

```python
final_data['Tổng'] = final_data.iloc[:, 1:].sum(axis=1)
```

```python
vi_tri = pd.read_excel('./tọa độ.xlsx')
```

```python
Ten_tram = vi_tri['Tên trạm'].to_list()
Toa_do = vi_tri['Tọa độ'].to_list()
Toa_do = [[float(value) for value in loc.split(', ')] for loc in Toa_do]
```

```python
districts = []

for i, tram in enumerate(Ten_tram):
    districts.append({"name": tram, "coord": Toa_do[i]})
```

```python
districts
```

```python
import folium
from folium.plugins import HeatMap

# Tạo bản đồ với tọa độ trung tâm tỉnh Đà nẵng
binh_phuoc_map = folium.Map(location=[16.067198, 108.223000], zoom_start=10)


rain = final_data['Tổng'].to_list()

for i, inf in enumerate(districts):
    inf['rainfall'] = rain[i]
    

# Tạo danh sách tọa độ và trọng số (lượng mưa) cho heatmap
heat_data = [[district["coord"][0], district["coord"][1], district["rainfall"]] for district in districts]

# Tạo Heatmap từ dữ liệu trên
HeatMap(heat_data).add_to(binh_phuoc_map)
binh_phuoc_map.add_child(folium.Circle([16.067198, 108.223000], radius = 31000, fill = True))
# Hiển thị bản đồ
binh_phuoc_map.save('binh_phuoc_rainfall_heatmap_day3.html')
```
