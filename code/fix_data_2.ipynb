```python
import numpy as np
import pandas as pd
import os
```

```python
files = os.listdir("../data")
```

```python
for file in files:
    df = pd.read_csv(f"../data/{file}")
    df = df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1)
    condition = (df[['pm25', 'pm10', 'o3', 'no2', 'so2', 'co']].sum(axis=1) == 0) & (df['aqi'] != 0)
    df.loc[condition, 'pm25'] = df['aqi']
    df.to_csv(f"../data/{file}")
```
